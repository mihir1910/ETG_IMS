# Purchasing Item from Inventory

```python
amt=0
product_id=prod.keys()
f = open("sale.json","a")
case = True
while case:
    prod_id = int(input("Please Enter Unique Id of the product that you want to buy: "))
    prod_quan = int(input("Enter the quantity of the product: "))
    count =count+ prod_quan
   
    
    f.write(str(prod[prod_id])+"Billing = "+str(prod[prod_id]["price"]*prod_quan))
    print("Total Bill of the Product: ", prod[prod_id]['price'] * prod_quan)
    print("Product: ", prod[prod_id]['name'])
    print("Amount of the Product: ", prod[prod_id]['price'])
    amt=amt+prod[prod_id]['price'] * prod_quan
    
    prod[prod_id]['quantity'] = prod[prod_id]['quantity'] - prod_quan
    presentime = datetime.now()

   
    print("Press 1 to proceed  else  any key to quit")
    prod_id= input("Enter your option: ")
    case = True if prod_id == '1'  else False
    
 

f.write("\nProduct sold = "+str(count)+"\n")
print('\n')
print("Total Bill of the Pruchase  for the customer is:", amt)
print("TIME OF PURCHASE:", presentime)
print("DATE OF PURCHASE:", presentime.strftime('%Y-%m-%d'))
print("THANKS FOR SHOPPING WITH US, PLEASE VISIT AGAIN😊😊")
f.close()
```
